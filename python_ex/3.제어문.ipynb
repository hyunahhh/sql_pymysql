```python
'''
입력받기 :  1<= 입력값 < 1000
input_data =,구분해서 split하기 => 리스트
input_data =[1, 20, 300, 3,4,5, 50, 500]

num_1 = []
num_2 = []
num_3 = []
for data in input_data:
    # 1자리수, 리스트에 넣기
    # 2자리수, 리스트에 넣기
    # 3자리수, 리스트에 넣기
# 각각 최종 출력
'''
```

```python
# input_data =['1', '20', '300', '3','4','5', '50', '500']
input_data = input("숫자를 ,로 구분해서 입력 예) 1,2,3,4  >> ")
input_data = input_data.split(",")
num_1 = []
num_2 = []
num_3 = []
for data in input_data:
    # 1자리수, 리스트에 넣기
    if len(data) == 1 :
        num_1.append(data)
    # 2자리수, 리스트에 넣기
    elif len(data) == 2 :
        num_2.append(data)
    # 3자리수, 리스트에 넣기
    elif len(data) == 3 :
        num_3.append(data)
# 각각 최종 출력
print(num_1)
print(num_2)
print(num_3)
```

# dict 확인문제2

```python
"""
# 결과- {숫자1:숫자1의 갯수, 숫자2:숫자2의 갯수, .... }

for 변수 in numbers:
    if 숫자가 처음 만나면,
        그 숫자가 키 : val=1
    처음 아니면:
        그 숫자가 키 : val+1
"""
```

```python
numbers = [1,2,6,8,4,3,2,1,9,7,2,1,3,5,4,8,9,7,2,3]
counter = {}
```

```python
for number in numbers:
    # if 숫자가 처음 만나면,
    if number not in counter:
        # 그 숫자가 키 : val=1
        counter[number]=1
        print(counter[number])
    # 처음 아니면:
    else:
        # 그 숫자가 키 : val+1
        counter[number]=counter[number]+1
print(counter)
```

```python
character = {
    "이름": "기사",
    "레벨": 12,
    "아이템": {
        "sward":"불꽃의 검",
        "armor":"풀플에이트",
    },
    "능력": ["베기", "세게 베기", "아주 세계 베기"]
}
character
```

```python
for key, val in character.items():
    # 값이 str 또는 int 인경우 처리
    if type(val) is str or type(val) is int:
        print(f"{key} : {val}")
    # val가 dict인 경우 출력
    if type(val) is dict:
        for sub_key, sub_val in val.items():
            print(f"{sub_key} : {sub_val}")
    # val가 list인 경우 출력
    if type(val) is list:
        for l_val in val:
            print(f"skill : {l_val}")
```

## while 확인문제2
- 1부터 숫자(i)를 1씩 증가시킴
- 누적 더해서 처음으로 10000을 넘기는 숫자(i) 구하기
- 최초로 누적합이 10000이 넘은 그 값이 얼마인지 구하기(10000 미포함)
- 숫자와 최초의 10000넘은 값 출력

```python
# 증가할 변수
i = 1
# 누적합 변수 초기화
sum_val = 0

limit = 10000
# limit = 20
while sum_val <= limit :  # 조건이 참일 동안 반복
    # i를 1씩 증가시키면서 누적 합(sum_val)
    sum_val = sum_val + i
    i = i + 1
print(f"{limit}을 넘는 i는 {i}이고, 합값은 {sum_val}입니다.")
```

# 확인문제3

- 1~100 : 사이의 소수구하기
- 1과 자기자신을 제외한 숫자로 나눠서 떨어지지 않으면(약수가 없음)
<pre>
- 2~101까지 숫자를 반복해서 각각 숫자가 소수인지 판단 (for)
    - 한 숫자(n)가 소수인지 판단
    - n/2~(n-1) 1씩줄이면서 나눴으때, 나눠지는 수가 없으면 소수
    - 나눠지는 수가 첫번째 발견되면 break
</pre>

```python
import
decimal = []
# 2 무조건 소수임
decimal.append(2)

# 3~100까지 소수판단
for n in range(2, 101):
    # 분모 : denominator(deno)
    deno = n-1 # 1과 자기 자신을 제외한 숫자로 나눔
    while (n/2) >= 2:
    # while (deno >= 2):
        # n = 6 이라면 판단방법
        if n % deno == 0:
            break
        if deno == 2:
            decimal.append(n)
        deno = deno -1
print(decimal)
```
