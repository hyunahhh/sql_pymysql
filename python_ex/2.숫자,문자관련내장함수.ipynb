# 문제2)
두 수 입력받아서 사칙연산, 나머지, 몫 구해서 출력

```python
nums = input("두수 입력 >> ")
print("입력 받은 값 : ", nums)
print("입력 받은 값의 타입 :", type(nums))
nums = nums.split(",")
print("스플릿 한 값 : ", nums)
print("스플릿 한 값의 타입 : ", type(nums))
num1 = eval(nums[0])
num2 = eval(nums[1])
print(f"{num1}  + {num2} = {num1 + num2}")
print(f"{num1}  - {num2} = {num1 - num2}")
print(f"{num1}  * {num2} = {num1 * num2}")
print(f"{num1}  % {num2} = {num1 % num2}")
print(f"{num1}  // {num2} = {num1 // num2}")
print(f"{num1}  ** {num2} = {num1 ** num2}")
```

# 문제3)
- 입력값이 숫자만, 문자로만, 숫자와문자로 이뤄져 있는지 판단하기
- 파이썬 조건문을 인터넷에서 검색해서 해결하기
- 문자만 있으면
  문자만 있습니다. 출력
- 숫자만 있으면
  숫자만 있습니다.
- 숫자,문자로 구성되어 있으면
  숫자 문자 모두 있습니다.

```python
# money = True
money = False
money
```

```python
if money :
  print("택시타기")
else:
  print("걷기")
```

```python
if money :
  print("택시타기")
if not(money) :
  print("걷기")
```

```python
# 대입연산자
money = 1000
# 비교연산자
print(money < 1000)
print(money > 1000)
print(money >= 1000)
print(money <= 1000)
print(money == 1000)
```

```python
money = 1000
if money > 4000:
  print("택시타기")
elif money > 2000 :
  print("지하철")
else:
  print("걷기")
```

```python
# 파이썬 조건문을 인터넷에서 검색해서 해결하기
# 문자만 있으면 문자만 있습니다. 출력
# 숫자만 있으면 숫자만 있습니다.
# 숫자,문자로 구성되어 있으면 숫자 문자 모두 있습니다.

# data = input()
# data = "aaaa1123"
# data = "aaaa"
# data = "1123"
data = "aaa 1234"

if data.isnumeric() :
  print("숫자만 있음")
elif data.isalpha() :
  print("문자만 있음")
elif data.isalnum():
  print("숫자, 문자")

```
